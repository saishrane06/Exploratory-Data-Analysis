```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
```

```python
df = pd.read_csv('ford.csv')
df.head()
```

```python
from sklearn import preprocessing
Transmission_temp = preprocessing.LabelEncoder()
Transmission_temp.fit(df["transmission"])
Transmission_temp.classes_
Transmission=Transmission_temp.transform(df["transmission"])


fueltype_temp =  preprocessing.LabelEncoder()
fueltype_temp.fit(df["fuelType"])
fueltype_temp.classes_
fueltype=fueltype_temp.transform(df["fuelType"])
```

```python
df.drop(["model","transmission","fuelType"],axis=1,inplace=True)
```

```python
Transmission=pd.Series(Transmission)
fueltype=pd.Series(fueltype)

df=pd.concat([df,Transmission,fueltype],axis=1)
df.rename(columns={"0":""})
df.head()
```

```python
df.rename(columns={0:'Transmission',1:'fueltype'},inplace=True)
df.columns
```

```python
df.columns
```

```python
X = df.drop('price',axis=1)
y = df['price']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
X_train.shape,X_test.shape
```

```python
var_threshold = VarianceThreshold(threshold=0)
var_threshold.fit(X_train)
```

```python
var_threshold.get_support()
```

```python
sum(var_threshold.get_support())
```

```python
len(X_train.columns[var_threshold.get_support()])
```

```python
constant_columns = [column for column in X_train.columns
                    if column not in X_train.columns[var_threshold.get_support()]]

print(len(constant_columns))
```
